```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('input/CONVENIENT_global_confirmed_cases.csv')
df_sum = pd.read_csv('input/country_wise_latest.csv',
                usecols = [0,1])
```

```python
df.head()
```

```python
df_sum.head()
df_sum = df_sum.sort_values(by=['Confirmed'],axis=0,ascending=False)
# 확진자수 Top10 나라
x_sum = df_sum['Country/Region'].tolist()[:10]
```

```python
df_sum
```

```python
# y값
y_top10 = []
for i in x_sum:
    y_top10.append(df[i][1:])
print(y_top10)
```

```python
# x는 날짜
x = df["Country/Region"][1:]
x_i = x.tolist()
```

```python
plt.figure(figsize=(30,30))
sns.lineplot(x_i,y_top10[0],label = x_sum[0])
sns.lineplot(x_i,y_top10[1],label = x_sum[1])
sns.lineplot(x_i,y_top10[2],label = x_sum[2])
sns.lineplot(x_i,y_top10[3],label = x_sum[3])
sns.lineplot(x_i,y_top10[4],label = x_sum[4])
sns.set(font_scale=2)
plt.xlim('1/23/20','2/15/20')
plt.xticks(rotation = '45')
plt.legend(loc = 1,fontsize='medium')
# plt.xlim(200,450)
```
