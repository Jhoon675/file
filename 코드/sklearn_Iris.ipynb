# iris 머신러닝 실습

```python
import sklearn
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import accuracy_score
```

```python
# 머신러닝에 필요한 iris data를 불러오기 
iris = load_iris()
# numpy의 ndarray로 변환
iris_data=iris.data
```

```python
iris
```

```python
# target = 결과값?
iris_label = iris.target
# print('iris target값:', iris_label)
# target명은 feature값
# print('iris target명:', iris.target_names)
```

```python
# feature_name은 불러온 iris data안에 포함되어 있다.
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
```

```python
iris_df['label'] = iris.target
```

```python
feature_names = iris_df.columns.tolist()[:-1]
```

```python
iris_df.head(3)
```

```python
# X_train: 학습 데이터, X_test = 결과 값(test를 하기 위해 보통의 경우에는 존재하지 않는다. \
#             y_train: 전체 데이터 중의 test 데이터의 비율, y_test: 항상 일정한 결과값을 얻기위해 랜덤값을 조정한다.)
X_train,X_test,y_train,y_test = train_test_split(iris_data, iris_label, test_size = 0.2, random_state=11)
```

## DecisionTreeClassifier 객체 생성

```python
dt_clf = DecisionTreeClassifier(random_state=11)
```

```python
# 학습 수행
dt_clf.fit(X_train, y_train)
```

```python
# 학습된 데이터로 예측 수행
pred = dt_clf.predict(X_test)
pred
```

```python
# accuracy_score : 정확도 측정 (실제 레이블 데이터 세트, 내가 학습 완료 시킨 예측 레이블 데이터 세트)
print('예측 정확도: {0:.4f}'.format(accuracy_score(y_test,pred)))
```
